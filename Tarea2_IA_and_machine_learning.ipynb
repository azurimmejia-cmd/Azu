## Declara dos numero enteros y realiza todas las operaciones aritméticas entre ellos

```python
n1=1
n2=20
print(f"La suma de los números es: {n1+n2}")
print(f"La resta de los números es: {n1-n2}")
print(f"La multiplicación de los números es: {n1*n2}")
print(f"La división de los números es: {n1/n2}")
print(f"El residuo de la división de los números es: {n1%n2}")
print(f"El cociente de la división de los números es: {n1//n2}")
print(f"El exponente de los números es: {n1**n2}")
```

#Programa que calcula el área de un círculo

```python

r=float(input("Ingresa el valor de tu radio :"))
u=input("En que unidades esta tu radio:")
pi=3.1416
area=pi*r**2
print(f"El área del círculo es: {area:.4f} {u}")
```

```python
n1=float(input("Ingresa el primer número: "))
if n1%2==0:
    print(f"El número {n1} es par")
else:
    print(f"El número {n1} es impar")
```

```python
# Ejemplo de operadores de asignación
x = 10
print(f"Valor inicial de x: {x}")

# Suma y asignación
x += 5
print(f"Después de x += 5: {x}")

# Resta y asignación
x -= 3
print(f"Después de x -= 3: {x}")

# Multiplicación y asignación
x *= 2
print(f"Después de x *= 2: {x}")

# División y asignación
x /= 4
print(f"Después de x /= 4: {x}")
```
