# **Programa que calcula el área de un rectángulo**

```python
b=float(input("Ingresa el valor de la base"))
h=float(input("Ingresa el valor de la altura"))
res=b*h
print(f"Hola el área de este rectangulo de acuerdo a tus datos es {res:.4f} ")
```

# Programa que calcula el promedio de 3 números

```python
v1=float(input("Ingresa tu primer calificación:"))
v2=float(input("Ingresa tu segunda calificación:"))
v3=float(input("Ingresa tu tercera calificación:"))
promedio=v1+v2+v3
print(f"Hola estimado usuario el promedio de tus caificaciones son :{promedio:.2f}")
```

```python
a0=float(input("Ingresa tu valor a convertir en grados celcius:"))
grados_f=a0*1.8+32
grados_r=a0*1.8+491.67
grados_k=a0+273.15
print(f"Tu valor en grados Fahrenheit  es: {grados_f:.3f}")
print(f"Tu valor en grados Rankie es: {grados_r:.4f}")
print(f"Tu valor en grados Kelvin es: {grados_k:.2f}")
```
