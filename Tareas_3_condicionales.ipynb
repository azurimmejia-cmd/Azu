```python
import random

# Generamos un número secreto aleatorio entre 1 y 100
numero_secreto = random.randint(1, 100)

intentos = 0
adivinado = False

while not adivinado:
    intentos += 1
    usuario = int(input("Adivina el número secreto (1-100): "))
    if intentos >= 5:
        print("Has agotado tus intentos. El número secreto era", numero_secreto)
        break

    if usuario == numero_secreto:
        print("¡Felicidades! Adivinaste el número en", intentos, "intentos.")
        adivinado = True
    elif usuario < numero_secreto:
        print("El número secreto es mayor.")
    else:
        print("El número secreto es menor.")
```
