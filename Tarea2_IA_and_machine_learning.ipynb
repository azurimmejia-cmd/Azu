````python
# Ejemplo de operadores de asignación
x = 10
print(f"Valor inicial de x: {x}")

# Suma y asignación
x += 5
print(f"Después de x += 5: {x}")

# Resta y asignación
x -= 3
print(f"Después de x -= 3: {x}")

# Multiplicación y asignación
x *= 2
print(f"Después de x *= 2: {x}")

# División y asignación
x /= 4
print(f"Después de x /= 4: {x}")
````

### Ejercio paraecido pero con ciclo

````python
#Este ejercicio no forma parte de los ejercicios de tarea sin embargo fue excepcional realizarlo ya que es un tema ligeramente más avanzado
s=100000 #Saldo iniclal

while True:
    a = float(input("Presiona 1 para depositar o 2 para retirar: "))

    if a == 1 or a == 2:
        break  # Sale del ciclo si la opción es válida
    else:
        print(" Error: ingresa una opción correcta (1 o 2).\n")
if a==1:
  q=float(input("Cuanto vas a depositara tu cuenta:"))
  s+=q
  print(f"Tu saldo actual es: {s}")
elif a==2:
  q=float(input("Cuanto vas a retirar de tu cuenta:"))
  s-=q
  print(f"Tu saldo actual es: {s}")
````
